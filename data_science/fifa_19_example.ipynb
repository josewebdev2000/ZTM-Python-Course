```python
# Try to grab the best and cheapest players
# Grab Players Names, Ages, Nationalities, Club, Wage, Potential, Overall, and Value

# Make a new data frame from the original one
df1 = pd.DataFrame(data_frame, columns=["Name", "Age", "Nationality", "Club", "Wage", "Potential", "Overall", "Value"])
df1
```

```python
# Enter function to convert str to float to get players difference by their value and wage
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
# Remove euro sign from columns wage and value with regex and covnert values to float
wage = df1["Wage"].replace('[\€,]', '', regex = True).apply(value_to_float)
value = df1["Value"].replace('[\€,]', '', regex = True).apply(value_to_float)
```

```python
wage # Show if it worked
```

```python
# Change values of original columns to the columns we processed
df1["Wage"] = wage
df1["Value"] = value
```

```python
df1.head()
```

```python
# Get difference between Value and Wage of players
df1["Difference"] = df1["Value"] - df1["Wage"]
```

```python
df1["Difference"]
```

```python
# Sort difference values
df1.sort_values("Difference", ascending = False)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
```

```python
# Make a graph that allows us to extract information from each point by hovering them
HOVER_DATA = [
    ("index", "$index"),
    ("Name", "@Name"),
    ("(Wage,Value)", "(@Wage, @Value)"),
    ("Club", "@Club"),
    ("Nationality", "@Nationality")
]

hover = HoverTool(tooltips = HOVER_DATA)

p = figure(title="FIFA 2019 Players", x_axis_label="Wage", y_axis_label="Value", plot_width=700, plot_height=700, tools = [hover])
p.circle('Wage', 'Value', size = 10, source = df1)
show(p)
```
